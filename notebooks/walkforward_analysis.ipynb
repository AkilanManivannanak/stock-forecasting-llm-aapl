```python
import os
os.getcwd()

import pandas as pd

path = "/Users/akilan/stock-forecasting-llm-aapl/walkforward_metrics.csv"
walk_df = pd.read_csv(path)
walk_df.head()

```

```python
summary = (
    walk_df
    .groupby(["regime", "model"])["rmse"]
    .mean()
    .reset_index()
    .pivot(index="regime", columns="model", values="rmse")
)
summary
```

### Walk-forward backtest interpretation

The table above shows the average RMSE for the naive and random forest models across different market regimes, computed from the walk-forward windows. In all regimes (bull, bear, and sideways), the naive baseline achieves a lower RMSE than the random forest model in this configuration. This indicates that, with the current features and hyperparameters, the random forest does not outperform the simple naive approach, and would require further tuning or additional features to add value over the baseline.
